```python
import json
import os

DATA_FILE = 'expenses.json'

def load_expenses():
    if not os.path.exists(DATA_FILE):
        return []
    with open(DATA_FILE, 'r') as file:
        return json.load(file)

def save_expenses(expenses):
    with open(DATA_FILE, 'w') as file:
        json.dump(expenses, file, indent=4)

def add_expense(expenses):
    date = input("Enter the date (YYYY-MM-DD): ")
    amount = float(input("Enter the amount: "))
    category = input("Enter the category (e.g., groceries, entertainment): ")
    description = input("Enter a description: ")

    expense = {
        "date": date,
        "amount": amount,
        "category": category,
        "description": description
    }
    expenses.append(expense)
    save_expenses(expenses)
    print("Expense added successfully.")

def list_expenses(expenses):
    if not expenses:
        print("No expenses found.")
        return

    for idx, expense in enumerate(expenses, start=1):
        print(f"{idx}. Date: {expense['date']}, Amount: {expense['amount']}, Category: {expense['category']}, Description: {expense['description']}")

def edit_expense(expenses):
    list_expenses(expenses)
    try:
        idx = int(input("Enter the expense number to edit: ")) - 1
        if 0 <= idx < len(expenses):
            expenses[idx]['date'] = input(f"Enter the new date (YYYY-MM-DD) [{expenses[idx]['date']}]: ") or expenses[idx]['date']
            expenses[idx]['amount'] = float(input(f"Enter the new amount [{expenses[idx]['amount']}]: ") or expenses[idx]['amount'])
            expenses[idx]['category'] = input(f"Enter the new category [{expenses[idx]['category']}]: ") or expenses[idx]['category']
            expenses[idx]['description'] = input(f"Enter the new description [{expenses[idx]['description']}]: ") or expenses[idx]['description']
            save_expenses(expenses)
            print("Expense updated successfully.")
        else:
            print("Invalid expense number.")
    except ValueError:
        print("Invalid input.")

def delete_expense(expenses):
    list_expenses(expenses)
    try:
        idx = int(input("Enter the expense number to delete: ")) - 1
        if 0 <= idx < len(expenses):
            expenses.pop(idx)
            save_expenses(expenses)
            print("Expense deleted successfully.")
        else:
            print("Invalid expense number.")
    except ValueError:
        print("Invalid input.")

def analyze_expenses(expenses):
    if not expenses:
        print("No expenses to analyze.")
        return

    total_spent = sum(expense['amount'] for expense in expenses)
    categories = {}
    for expense in expenses:
        categories[expense['category']] = categories.get(expense['category'], 0) + expense['amount']

    print(f"\nTotal spent: {total_spent}")
    print("Spending by category:")
    for category, amount in categories.items():
        print(f"{category}: {amount}")

def main():
    expenses = load_expenses()

    while True:
        print("\nExpense Tracker")
        print("1. Add expense")
        print("2. List expenses")
        print("3. Edit expense")
        print("4. Delete expense")
        print("5. Analyze expenses")
        print("6. Exit")

        choice = input("Choose an option: ")

        if choice == '1':
            add_expense(expenses)
        elif choice == '2':
            list_expenses(expenses)
        elif choice == '3':
            edit_expense(expenses)
        elif choice == '4':
            delete_expense(expenses)
        elif choice == '5':
            analyze_expenses(expenses)
        elif choice == '6':
            print("Exiting the application.")
            break
        else:
            print("Invalid choice. Please choose a valid option.")

if __name__ == "__main__":
    main()
```
